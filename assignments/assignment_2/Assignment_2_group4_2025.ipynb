- Use a combination of `for and while` loops to calculate a rolling cumulative revenue:

For each category, use a for loop to iterate over its transactions.
Use a while loop to stop accumulating revenue for a category once it exceeds $ 3000.

Print the cumulative revenue for each category and the number of transactions it took to exceed $3000.

```python
#Creamos el bucle para iterar sobre cada categoría de transacciones. #
for category in sales_data["Category"].unique():
    category_sales = sales_data[sales_data["Category"] == category]
    accumulated_revenue = 0
    transaction_count = 0
    iterator = iter(category_sales.iterrows())
#Usamos un while loop para acumular los ingresos hasta que superen los $3000. #
    while accumulated_revenue <= 3000:
        try:
            _, row = next(iterator) #Usamos un iterador con next() para recorrer las transacciones una por una #
            accumulated_revenue += row["Total"]
            transaction_count += 1
        except StopIteration:
            break #Detenemos el ciclo while cuando superamos los $3000 o cuando no quedan más transacciones.#
    print(f"\nCategoría {category}: ${accumulated_revenue:} acumulados en {transaction_count} transacciones.")
#Finalmente, imprimimos el total acumulado y la cantidad de transacciones utilizadas.#
```

1. You are tasked with calculating the average grade for each student in the dataset. However, the dataset has some issues:

Some scores are marked as "missing" or "invalid" (strings instead of numeric values).
Some students may have incomplete grades across subjects.

To calculate the average:

Skip any "invalid" or "missing" values during calculations.
If a student has no valid grades, print a warning message for that student.


Use `try` and `except` to handle potential errors.
Use `continue` to skip invalid rows or scores.
Print the results in a clear format, showing the average grade for each student or a warning if no grades are available.

```python
#importamos la base de datos "students_data"#
students_data= pd.read_csv("C:/Users/Santiago/Documents/GitHub/summer_python_medialab/assignments/assignment_2/students_grade.csv")
students_data
```

```python
# creamos un diccionario para almacenar promedios
student_averages = {}
```

```python
# creamos la función para calcular el promedio ignorando valores "missing" o "invalid"
def calculate_average(row):
    valid_grades = []
    
    for subject in ["Math", "English", "Science", "History"]:
        try:
            grade = float(row[subject])  # Intentamos convertir la calificación a número
            valid_grades.append(grade)
        except ValueError:  # utilizamos  except para ignorarlo si no se puede convertir 
            print(f"⚠ Warning: Invalid grade for {row['Name']} in {subject}")
            continue

    # Utilizamos If para mostramos advertencia en caso no existan calificaciones válidas
    if not valid_grades:
        print(f"⚠ Warning: No valid grades for {row['Name']}")
        return None
    #finalmente calcuulamos el promedio
    return round(sum(valid_grades) / len(valid_grades), 2)  
```

```python
# Aplicamos la función a cada estudiante
students_data["Average"] = students_data.apply(calculate_average, axis=1)

# Mostramos los resultados
print(students_data[["Student_ID","Name","Average"]])
```

2. Calculate the average grade per subject for each year.
Handle invalid or missing values during the calculation by skipping them.
Print the averages in a clean format (e.g., table or grouped output).

```python
#Convertimos las columnas de materias a valores numéricos.


def calculate_subject_average(df):
    subjects = ["Math", "English", "Science", "History"]
    subject_averages = {}
    for subject in subjects:
        df[subject] = pd.to_numeric(df[subject], errors='coerce')  # Convertimos a numérico, ignorando errores
        subject_averages[subject] = df.groupby("Year")[subject].mean().round(2)
        #Devolvemos un DataFrame con los resultados.
    return pd.DataFrame(subject_averages)
```

```python
# Calculamos los promedios
subject_avg_per_year = calculate_subject_average(students_data)

# Mostramos los resultados
print("\nPromedio de calificaciones por materia y año")
print(subject_avg_per_year)
```
